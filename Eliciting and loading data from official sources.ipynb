# Preliminaries

```python
import pandas as pd
import numpy as np
from pathlib import Path

data_dir = Path("data")

def get_dataframe(name):
    path = data_dir / f"{name}.csv"
    
    if not path.exists():
        raise FileNotFoundError()
    
    return pd.read_csv(path)
```

# Confirmed Global Cases

```python
confirmed_df = get_dataframe('confirmed_df')
confirmed_df
```

```python
confirmed_df.describe()
```
